```python
import json
import pandas as pd
import numpy as np
import os
```

```python
# read json file
with open('./data_qna (3).json') as f:
    data = json.load(f)

keys = list(data.keys())
values = list(data.values())
df = pd.DataFrame()
for key in keys:
    print(key, data[key]['questions'])
    leng = len(data[key]['questions'])
    df = pd.concat([df, pd.DataFrame({'question': data[key]['questions'], 'topic': [key]*leng})], ignore_index=True)
    
df.head()
df.to_csv('./data_qna.csv', index=False, encoding='utf-8-sig')
```
